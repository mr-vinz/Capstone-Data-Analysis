### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
##Write your code here
sns.scatterplot(data=df,x="Age",y="SOPartFreq",size="JobSat",legend=False,sizes=(20, 2000),alpha=0.5)
plt.tick_params(axis='x', rotation=90)
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
replacements = {'Under 18 years old': 18, '35-44 years old': 39.5,'45-54 years old': 49.5,'18-24 years old': 21,'25-34 years old': 29.5,'55-64 years old': 59.5,'65 years or older': 65,'Prefer not to say':0}
df['Age'] = df['Age'].replace(replacements)
```

```python
##Write your code here
sns.scatterplot(data=df,x="JobSat",y="ConvertedCompYearly",size="Age",legend=False,sizes=(20, 2000),alpha=0.5)

plt.show()
```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
frequency_df=df[df['LanguageHaveWorkedWith'].isin(df['LanguageHaveWorkedWith'].dropna().value_counts().keys()[0:10])]
```

```python
frequency_df = frequency_df.groupby(['LanguageHaveWorkedWith', 'Age']).size().reset_index(name='Frequency')
```

```python
##Write your code here
sns.scatterplot(data=frequency_df ,x='LanguageHaveWorkedWith',y='Age',size='Frequency',legend=False,sizes=(20, 2000),alpha=0.5)
plt.tick_params(axis='x', rotation=90)
plt.show()
```

#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.


```python
frequency_df=df[df['DatabaseWantToWorkWith'].isin(df['DatabaseWantToWorkWith'].dropna().value_counts().keys()[0:10])]
frequency_df = frequency_df.groupby(['DatabaseWantToWorkWith', 'JobSat']).size().reset_index(name='Frequency')
```

```python
##Write your code here
sns.scatterplot(data=frequency_df ,x='DatabaseWantToWorkWith',y='JobSat',size='Frequency',legend=False,sizes=(20, 2000),alpha=0.5)
plt.tick_params(axis='x', rotation=90)
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
frequency_df=df[df['WebframeWantToWorkWith'].isin(df['WebframeWantToWorkWith'].dropna().value_counts().keys()[0:10])]
frequency_df = frequency_df.groupby(['WebframeWantToWorkWith', 'JobSat']).size().reset_index(name='Frequency')
```

```python
##Write your code here
sns.scatterplot(data=frequency_df ,x='WebframeWantToWorkWith',y='JobSat',size='Frequency',legend=False,sizes=(20, 2000),alpha=0.5)
plt.tick_params(axis='x', rotation=90)
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
frequency_df=df[df['LanguageAdmired'].isin(df['LanguageAdmired'].dropna().value_counts().keys()[0:10]) & df['Country'].isin(df['Country'].dropna().value_counts().keys()[0:10])]
frequency_df = frequency_df.groupby(['LanguageAdmired', 'Country']).size().reset_index(name='Frequency')
```
